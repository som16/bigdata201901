```python
# 인터넷 연결
# html 문서를 받아와서 분석

import requests
from bs4 import BeautifulSoup
```

```python
url = "https://finance.naver.com/item/main.nhn?code=035420"
```

```python
result = requests.get(url)

result
# 코드연결 확인
```

```python
bs_obj = BeautifulSoup (result.content, "html.parser")
```
